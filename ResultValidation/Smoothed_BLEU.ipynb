```python
from nltk.translate.bleu_score import sentence_bleu
import pandas as pd
```

```python
def get_bleu_score(reference,candidate):
      smoothie = SmoothingFunction().method4
      score = sentence_bleu(reference.split(), candidate.split(),smoothing_function=smoothie)
      return score

reference = "allo testing"
candidate = "hello testing"
print(get_bleu_score(reference,candidate))

def get_bleu_score_record(x):
  return get_bleu_score(x['Expected'],x['predictions'])


def get_average_bleu_score(filename,sample_frac=1):
  df = pd.read_json(filename,lines=True).sample(frac=sample_frac)
  print(len(df))
  df['bleu']=df.apply(get_bleu_score_record,axis=1)
  meanscore=df['bleu'].mean()
  print(f"Average BLEU Score for the model:{meanscore}")
  return meanscore
```
